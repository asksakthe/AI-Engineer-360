```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df1 = pd.read_csv('country_wise_latest.csv')
df2 = df1.loc[:,['Confirmed','New cases']]
```

```python
#Compute Statistical Measures
#mean
print(f"Mean Value of given DataSet: \n\t{df2.mean()}\n")
print(f"median Value of given DataSet: \n\t{df2.median()}\n")
print(f"variance Value of given DataSet: \n\t{df2.var()}\n")
print(f"Standard deviation Value of given DataSet: \n\t{df2.std()}\n")
print(f"Corelation Value of given DataSet: \n\t{df2.corr()}")
```

```python
#Outliers on Confirmed Feature
Q1 = df2['Confirmed'].quantile(.25)
Q3 = df2['Confirmed'].quantile(.75)

IQR = Q3-Q1

lower_bound = Q1 - (1.5* IQR)
upper_bound = Q3 + (1.5* IQR)
#print(lower_bound, upper_bound)
outlier_Confirmed = df2[(df2['Confirmed']> lower_bound) | (df2['Confirmed']<upper_bound)]


```

```python
#Outliers on New Cases Feature
Qn1 = df2.quantile(.25)
Qn3 = df2.quantile(.75)
IQRn = Qn3-Qn1

low = Qn1 - (1.5*IQRn)
high =Qn3+ (1.5*IQRn)

outlier_NewCases = df2[(df2['New cases'])>low | (df2['New cases']<high)]
```
