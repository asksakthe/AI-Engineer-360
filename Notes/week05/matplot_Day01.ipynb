```python
#Visualizing Test Results with Matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.title("Test Execution Results")
x = ['Passed','Failed','Skipped']
y = [45,10,5]

plt.bar(x,y, width=0.5, color = 'orange')
plt.xlabel("Test Status")
plt.ylabel("Number of Test Cases")

plt.show()
```

```python
#Defect Trend Analysis with Matplotlib
Weeks = [1, 2, 3, 4, 5, 6]
Defects = [5, 8, 6, 10, 7, 4]

plt.title("Defect Trend Over Time")
plt.xlabel("Week Number")
plt.ylabel("Number of Defects")

plt.plot(Weeks, Defects, marker='D', color= 'g', linestyle='-.')
plt.grid(False)
plt.savefig('chart_Line.png')
plt.show()



```

```python
#Pie Chart  Defect Distribution
x = ['high','medium','low']
y = [10,15,5]
plt.pie(y, labels=x, autopct='%1.1f%%',startangle=270)
plt.title("Defect Distribution by Severity")
plt.legend('sample',loc='lower right')
plt.show()

```

```python
data01 = [12, 15, 20, 18, 22, 30, 25, 16, 19, 28, 24, 14]
plt.xlabel = "Duration (seconds)"
plt.ylabel = "Number of Test Cases"
plt.title("Distribution of Test Execution Times")

plt.hist(data01, bins=5, color='grey', edgecolor='black')
plt.show()
```
