```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SalesDataSet.csv')
print(df.columns)
print(df.describe())
```

```python
# What is the 25th percentile of the Total Amount?
print("25th Percentile of Total Amount:",df['Total Amount'].quantile(0.25))
# What is the median (50th percentile) of the Total Amount?
print("Median (50th Percentile) of Total Amount:",df['Total Amount'].median())
# What is the 75th percentile of the Total Amount?
print("75th Percentile of Total Amount:",df['Total Amount'].quantile(0.75))
# What is the variance in Total Amount across all transactions?
print("*****Mean in Total Amount:",df['Total Amount'].mean())
print("******Variance in Total Amount:",df['Total Amount'].var())
# What is the variance in Quantity sold?
print("Variance in Quantity Sold:",df['Quantity'].var())
# What is the correlation between Age and Total Amount?
c01 = df[['Age','Total Amount']].corr()
print(f"Correlation between Age and Total Amount: {c01} and {df['Age'].corr(df['Total Amount'])}")
# What is the correlation between Quantity and Total Amount?
print("Correlation between Quantity and Total Amount:",df['Quantity'].corr(df['Total Amount']))
# What is the correlation between Price per Unit and Total Amount?
print("Correlation between Price per Unit and Total Amount:",df['Price per Unit'].corr(df['Total Amount']))


```

```python
df_new = df[['Age', 'Quantity', 'Price per Unit', 'Total Amount']]
print(df_new.corr())
print(plt.imshow(df_new.corr()))
plt.colorbar()
```
