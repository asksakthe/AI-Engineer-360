For the give data set using IQR technique  a) find out the outliers and write it to CSV b) write the cleaned data to CSV by excludig the outliers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df01 = pd.read_csv('SalesDataset-2.csv')
df01.head()
a = 1 
b = 5
c = 3
print((a < b) or (a > c))
```

```python
q1 = df01['Total Amount'].quantile(.25)
q3 = df01['Total Amount'].quantile(.75)
IQR = q3-q1

lower = q1-(1.5*IQR)
upper = q3+(1.5*IQR)

out = df01[(df01['Total Amount'] < lower) | (df01['Total Amount'] > upper)]
#outlier save in csv

out.to_csv('with_outlier.csv', index=False)
print(out.index)
```

```python
remove_out_DF = df01.drop(out.index)
remove_out_DF.to_csv('outlier_Removed_File.csv', index=False)
```
