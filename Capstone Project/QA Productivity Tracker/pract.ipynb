```python
def IQR_Outlier_Detect(col_name):
    Q1 = df[col_name].quantile(.25)
    Q3 = df[col_name].quantile(.75)
    IQR = Q3-Q1
    lower = Q1 - (1.5*IQR)
    upper = Q3 + (1.5*IQR)

    outlier = df[col_name] <lower | df[col_name] > upper
    if outlier:
        print("outliers are detected")
        return outlier
    else:
        print("NO outliers are detected")

    df
```
